Filtering Data Query Functions

```python
import pandas as pd
bios=pd.read_csv("../data/bios.csv")
```

```python
#1. Find athletes born in the USA using query():
bios.query('born_country=="USA"')
```

```python
# 2. Find athletes born in the USA and in the city of Seattle using query():
bios.query('born_country=="USA" and born_city=="Seattle"')
```
