```python
coffee=pd.read_csv("../warmup-data/coffee.csv")
coffee
```

```python
coffee['price']=4.99
coffee['new_price']=np.where(coffee['Coffee Type']=='Espresso',3.99,5.99)
coffee
```

```python
# 12. Reordering the columns in the coffee DataFrame to have 'Day', 'Coffee Type', and 'Units Sold' as the first three columns using column indexing and modifying the original DataFrame
coffee=coffee[['Day','Coffee Type','Units Sold','new_price']]
coffee
```

```python
# 13. Adding a new column 'revenue' that calculates the total revenue for each row by multiplying 'Units Sold' with 'new_price' and modifying the original coffee DataFrame
coffee['revenue']=coffee['Units Sold']*coffee['new_price']
coffee
```

```python
# 14. Renaming the 'new_price' column to 'Price' and the 'revenue' column to 'Revenue' in the coffee DataFrame using the rename() method and modifying the original DataFrame
# coffee=coffee.rename(columns={'new_price':'Price'})
coffee.rename(columns={'new_price':'Price'},inplace=True)
coffee.rename(columns={'revenue':'Revenue'},inplace=True)
coffee
```
